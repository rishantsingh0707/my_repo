### CRUD Operations

#### CRUD stands for Creating, Reading, Updating and Deleting the Table values

Out of which we have already learned to create a table. Now the rest of the CRUD operations are:

#### INSERTING DATA INTO THE TABLE
The keyword INSERT, which is used to insert new data/values in the table. When a table is created, the first operation after it is inserting fresh/new data into the table. 
#### UPDATING DATA INTO THE TABLE
The keyword UPDATE, which is used to update/modify existing data into the table
#### DELETING DATA FROM THE TABLE
The keyword DELETE is used to delete particular data/row from the table. DELETE is a DML (Data Manipulation Language) command that removes one or more rows from a table based on a condition.
 
One more command that is used to remove data from the table is TRUNCATE. It is a DDL (Data Definition Language) command that removes all rows from a table, without logging individual row deletions. It is faster and uses less transaction log space.

Other than CRUD operations, we also have the accessibility to alter columns in the table using ALTER function

### JOINS
<strong>Joins are used to combine rows from two or more tables based on a related/similar column between them.</strong>

<strong>Join Types:</strong> Inner Join, Left Join, Right Join, Full Join

<img src= "JOIN-TYPES.jpg" width="820"  alt="Join types">
<img src= "SQL-JOIN-TYPES.png" width="820"  alt="Join types">

```python
"""INNER JOIN: Inner Joins output only the rows that have matching values in both tables.

SYNTAX: SELECT column_name(s) FROM table1 INNER JOIN table2 ON table1.column_name = table2.column_name
    where column_name is the name of the column that is common between the two tables.
"""

query = """
        SELECT * FROM EmployeeDemographics INNER JOIN EmployeeSalary
        ON EmployeeDemographics.EmployeeID = EmployeeSalary.EmployeeID
        """
# query =  """SELECT EmployeeDemographics.EmployeeID, EmployeeDemographics.FirstName, 
#             EmployeeSalary.JobTitle, EmployeeSalary.Salary 
#             FROM EmployeeDemographics INNER JOIN EmployeeSalary 
#             ON EmployeeDemographics.EmployeeID = EmployeeSalary.EmployeeID"""


cursor.execute(query)

data = cursor.fetchall()
for d in data:
    print(d)
```

```python
"""
LEFT JOIN : Left Joins output all the rows from the left table and the matching rows from the right table.

SYBTAX: SELECT column_name(s) FROM table1 LEFT JOIN table2 ON table1.column_name = table2.column_name.

Example:Consider the two tables EmployeeDemographics and EmployeeSalary, and the query below.
        
        SELECT * FROM EmployeeDemographics LEFT JOIN EmployeeSalary 
        ON EmployeeDemographics.EmployeeID = EmployeeSalary.EmployeeID
        
        Here, if we want to get all the rows from the EmployeeDemographics table and only the matching rows 
        from the EmployeeSalary table, we can use the LEFT JOIN clause.
"""

query = """
        SELECT * FROM EmployeeDemographics LEFT OUTER JOIN EmployeeSalary
        ON EmployeeDemographics.EmployeeID = EmployeeSalary.EmployeeID
        """
cursor.execute(query)
data = cursor.fetchall()

for d in data:
    print(d)
```

```python
"""
RIGHT JOIN: Outputs all the rows from the right table and the matching rows from the left table.

SYNTAX: SELECT column_name(s) FROM table1 RIGHT JOIN table2 ON table1.column_name = table2.column_name.

Example: Consider the two tables EmployeeDemographics and EmployeeSalary, and the query below.
        
        SELECT * FROM EmployeeDemographics RIGHT JOIN EmployeeSalary 
        ON EmployeeDemographics.EmployeeID = EmployeeSalary.EmployeeID
        
        Here, if we want to get all the rows from the EmployeeSalary table and only the matching rows 
        from the EmployeeDemographics table, we can use the RIGHT JOIN clause.
"""

query = """SELECT EmployeeSalary.EmployeeID, FirstName, LastName,
            JobTitle, Salary 
            FROM EmployeeDemographics RIGHT JOIN EmployeeSalary
            ON EmployeeDemographics.EmployeeID = EmployeeSalary.EmployeeID"""
cursor.execute(query)
data = cursor.fetchall()

for d in data:
    print(d)
```

```python
"""
OUTER JOIN: Outputs all the rows from both tables, and also the rows that do not have matching values. 

SYNTAX: SELECT column_name(s) 
        FROM table1 FULL OUTER JOIN table2 ON table1.column_name = table2.column_name.

Example: Consider the two tables EmployeeDemographics and EmployeeSalary, and the query below.
        
        SELECT * FROM EmployeeDemographics FULL OUTER JOIN EmployeeSalary 
        ON EmployeeDemographics.EmployeeID = EmployeeSalary.EmployeeID
        
        Here, if we want to get all the rows from both tables, we can use the FULL OUTER JOIN clause.
"""
query = """SELECT * FROM EmployeeDemographics FULL OUTER JOIN EmployeeSalary
            ON EmployeeDemographics.EmployeeID = EmployeeSalary.EmployeeID"""
cursor.execute(query)
data = cursor.fetchall()

for d in data:
    print(d)
```

```python
"""
Likewise, if we want to get the data from table1 (left table) but only those data rows from table2 
(right table) that do not have matching values on column_name, we could eventually use LEFT OUTER JOIN.
"""

query = """SELECT * FROM EmployeeDemographics LEFT OUTER JOIN EmployeeSalary
            ON EmployeeDemographics.EmployeeID = EmployeeSalary.EmployeeID"""   
cursor.execute(query)
data = cursor.fetchall()

for d in data:
    print(d)
```

### Having Clause
The WHERE and HAVING clauses are used to filter data in SQL queries. Here are some difference between the two.

<strong>WHERE Clause:</strong> The WHERE clause is used to filter data from specific rows (individual rows) from a table based on certain conditions. It implements row operations and cannot contain an aggregate function. It can be used with a single row function and is applied before any groupings are made.

<strong>HAVING Clause:</strong> The HAVING clause is used to filter values from a group and can contain an aggregate function. It implements column operations and is used with a multiple-row function. It is applied after the GROUP BY clause and is used to filter grouped data.

#### <strong>Guidelines on when to use them and when not to:</strong>

##### WHERE Clause

<strong>Use WHERE:</strong>

Filtering individual rows: Use the WHERE clause to filter individual rows based on conditions such as equality, inequality, or range checks.

Pre-aggregation filtering: Use the WHERE clause to filter data before applying aggregate functions or grouping.

Non-aggregate conditions: Use the WHERE clause for conditions that dont involve aggregate functions, such as checking for NULL or non-NULL values.

<strong>Dont Use WHERE:</strong>

Aggregate conditions: Dont use the WHERE clause for conditions that involve aggregate functions, such as SUM, AVG, or COUNT.

Post-aggregation filtering: Dont use the WHERE clause to filter data after applying aggregate functions or grouping. Instead, use the HAVING clause.

##### HAVING Clause
<strong>Use HAVING:</strong>

Filtering grouped data: Use the HAVING clause to filter grouped data based on conditions such as aggregate functions or grouping criteria.

Post-aggregation filtering: Use the HAVING clause to filter data after applying aggregate functions or grouping.

Aggregate conditions: Use the HAVING clause for conditions that involve aggregate functions, such as checking if the sum of a column is greater than a certain value.

<strong>Dont Use HAVING:</strong>

Individual row filtering: Dont use the HAVING clause to filter individual rows. Instead, use the WHERE clause.

Non-aggregate conditions: Dont use the HAVING clause for conditions that dont involve aggregate functions.

________________________________________________________________

```python
# Basically, we use where clause to filter data before it is grouped/when row level filtering is needed.
# Having clauee is used after grouping & aggregation to filter the data in columns/in the entire column.

# Example of using where clause to filter data before grouping and aggregation
# query = """SELECT JobTitle, COUNT(JobTitle) 
#             FROM EmployeeSalary INNER JOIN EmployeeDemographics
#             on EmployeeSalary.EmployeeID = EmployeeDemographics.EmployeeID
#             WHERE COUNT(JobTitle) > 1
#             GROUP BY JobTitle
#             """

# OUTPUT FOR THIS QUERY WILL BE:
# ---------------------------------------------------------------------------
# OperationalError: misuse of aggregate: COUNT()


# Example of using having clause to filter data after grouping and aggregation
query = """SELECT JobTitle, COUNT(JobTitle) AS JobTitleCount, AVG(Salary) AS AverageSalary 
            FROM EmployeeSalary FULL OUTER JOIN EmployeeDemographics
            ON EmployeeSalary.EmployeeID = EmployeeDemographics.EmployeeID
            GROUP BY JobTitle
            HAVING AVG(Salary) > 25000
            ORDER BY AVG(Salary) DESC
            """
cursor.execute(query)
data = cursor.fetchall()
for d in data:
    print(d)
```

### Aliasing

Aliasing is used to give a table, or a column in a table, a temporary name. 

Example, consider the following query:

<code>
query = '''SELECT EmployeeDemographics.EmployeeID, EmployeeDemographics.FirstName, EmployeeSalary.Salary<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;FROM EmployeeDemographics FULL OUTER JOIN EmployeeSalary<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;ON EmployeeDemographics.EmployeeID = EmployeeSalary.EmployeeID'''<br>
</code>

In the above query, EmployeeDemographics and EmployeeSalary are the table names.

We can give them alias names as shown below:

<code>
query = '''SELECT E.EmployeeID, E.FirstName, S.Salary<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;FROM EmployeeDemographics AS E FULL OUTER JOIN EmployeeSalary AS S<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;ON E.EmployeeID = S.EmployeeID'''<br>
</code>

```python
query =  """SELECT sal.JobTitle, COUNT(JobTitle) AS JobTitleCount, AVG(Salary) AS AvgSal
            FROM EmployeeSalary sal INNER JOIN EmployeeDemographics demo
            ON sal.EmployeeID = demo.EmployeeID
            GROUP BY JobTitle
            HAVING AVG(Salary) > 25000
            ORDER BY AVG(Salary) DESC
            """
cursor.execute(query)
data = cursor.fetchall()
for d in data:
    print(d)
```

### PARTITION BY
 
PARTITION BY clause is used to divide the results into partitions to which the window function is applied

PARTITION BY is different from GROUPBY in the sense that GROUP BY combines all rows into a single group,
whereas PARTITION BY divides the rows into groups.

Example:<br>
Consider the EmployeeSalary table. We want to calculate the average salary for each job title.<br>
We can use the PARTITION BY clause to divide the result set into partitions based on the job title and then calculate the average salary for each partition<br>

How GROUPBY and PARTITION BY are different?

GROUP BY: Combines all rows into a single group
PARTITION BY: Divides the rows into groups

Example query using PARTITION BY:
<code><br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;SELECT EmployeeID, JobTitle, Salary,<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;AVG(Salary) OVER(PARTITION BY JobTitle) AS AvgSalary<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;FROM EmployeeSalary <br>
</code>

Here, the clause OVER(PARTITION BY JobTitle) divides the result set into partitions based on the 
job title and then calculates the average salary for each partition.

Example query using GROUP BY:
<code><br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;SELECT JobTitle, AVG(Salary) AS AvgSalary<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;FROM EmployeeSalary<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;GROUP BY JobTitle<br>
</code>

Here, the GROUP BY clause combines all rows into a single group and calculates the average salary 
for each job title.

So which one to use?

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;If you want to calculate the average salary for each job title, use PARTITION BY.<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;If you want to calculate the average salary for all job titles combined, use GROUP BY.<br>

Although we can use GROUP BY clause to calculate the average salary for each job title, 
like in the example below:
<code><br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;SELECT sal.JobTitle, COUNT(JobTitle) AS JobTitleCount, AVG(Salary) AS AvgSal<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;FROM EmployeeSalary sal INNER JOIN EmployeeDemographics demo<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;ON sal.EmployeeID = demo.EmployeeID<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;GROUP BY JobTitle<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;ORDER BY AVG(Salary) DESC<br>
</code>

The query above will give us the average salary for each job title, but at the cost of 
query complication and length of query.
    
Using PARTITION BY clause is more efficient and easier to understand.

```python
# Example of PARTITION BY clause. In this example,we will use two talbes 
# EmployeeDemographics and EmployeeSalary and will partition the data based on Gender and Salary

query = """SELECT Demo.FirstName, LastName, Gender, Salary,
            COUNT(Gender) OVER(PARTITION BY Gender) AS GenderCount 
            FROM EmployeeDemographics Demo FULL OUTER JOIN EmployeeSalary Sal
            ON Demo.EmployeeID = Sal.EmployeeID
            
"""
cursor.execute(query)
data = cursor.fetchall()
for d in data:
    print(d)

# In this example, the data is partitioned based on Gender after calculating the count of each gender.
# In the output data, we can see the total count of each gender in the GenderCount column.
# So when showing the data, if the employee is suppose, Female, then the GenderCount will be the 
# total count of Female employees in the table and likewise for Male employees.
```

### CTEs (Common Table Expressions)

CTEs (Common Table Expressions) are temporary result sets that you can reference within a SELECT, INSERT, UPDATE, or DELETE statement.

The WITH clause is used to define a CTE. The WITH clause can be used to define one or more CTEs.

It is a way to create a temporary table copy that can be used in the query.

When a CTE is created, it is only available to the query that created it. After that, if we try to access the CTE in another query, it will not be available.

```python
query = """WITH CTE_Employee AS 
            (SELECT Demo.EmployeeID, FirstName, LastName, JobTitle, Gender, Salary,
            COUNT(Gender) OVER(PARTITION BY Gender) AS GenderCount,
            AVG(Salary) OVER(PARTITION BY JobTitle) AS AvgSalary
            FROM EmployeeDemographics Demo JOIN EmployeeSalary Sal
            ON Demo.EmployeeID = Sal.EmployeeID
            )
            SELECT * FROM CTE_Employee
            """
cursor.execute(query)
data = cursor.fetchall()
for d in data:
    print(d)
```
